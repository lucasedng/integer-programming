```python
# Funcoes Auxiliares

# Funcao pertence verifica se o elemento x esta no vetor v
def pertence(x, v):
    for i in range(v.size):
        if(v[i] == x):
            return True
    return False

# Funcao NN retorna o indice da coluna da matriz M cujo elemento na linha A tem o menor valor e a coluna nao esteja no vetor v


def NN(A, M, v):
    r = inf
    p = A
    for j in range(M[A].size):
        if(M[A][j] < r and not(pertence(j, v))):
            p = j
            r = M[A][j]
    return p

# Funcao Medio retorna o vertice da matriz M com menor posicao media entre todos os vertices em v


def Medio(M, v):
    r = inf
    prox = 0

    if(caminho.size == 1):
        prox = 0
        for j in range(M[0].size):
            if(M[0][j] < r and not(pertence(j, v))):
                prox = j
                r = M[0][j]
        return prox

    for j in range(M[0]. size):
        T = 0
        if(not(pertence(j, v))):
            for i in range(M[0].size):
                if(pertence(i, v)):
                    T = T + M[i][j]
            T = T/(M[0].size + 1)
        if(T <= r and T != 0):
            prox = j
            r = T
    return prox

# Funcao melhor ve qual a melhor posicao para deixar x em v de acordo com a matriz M


def Melhor(x, M, v):
    melhor = inf
    coord = v.size
    for i in range(v.size+1):
        testar = np.insert(v, i, x)
        T = Total(M, testar)
        if(T <= melhor):
            melhor = T
            coord = i
    return coord

# Funcao Total calcula o tamanho total do caminho v de acordo com a matriz de distancias M


def Total(M, v):
    total = 0
    for i in range(v. size-1):
        total = total + M[v[i]][v[i+1]]
    total = total + M[v[0]][v[v.size-1]]
    return total

# Funcao Diferenca imprime a diferenca do total do caminho otimo para o caminho v encontrado pela heuristica (precisa mudar o nome do arquivo)


def Diferenca(M, v):
    sol = np.loadtxt("Sol_26.txt").astype(int) - 1
    n = sol.size - 1

    min = 0
    total = 0
    for i in range(n):
        min = min + M[sol[i]][sol[i + 1]]
        total = total + M[v[i]][v[i+1]]
    print('min =', min, 'heuristica = ', total, 'diferenca = ', total - min)
```
